```python
url = 'https://www.alamy.com/search.html?qt=hyssopus%20officinalis&qt_raw=hyssopus%20officinalis&ps=673'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
images = soup.findAll('img')
urls1=[]
for image in images:
    name = image.get('alt')
    link = image.get('src')
    urls1.append(link)
print(urls1)
```

```python
urls1 = [x for x in urls1 if x != None]
urls1
```

```python
urls1 = [b.replace('/6/','/3/') for b in urls1]
urls1
```

```python
import urllib.request
```

```python
for index, url in enumerate(urls1):
    link = urllib.request.urlopen(url)
    try:
        name = "sla%s.jpg" % (index+1)
        with open(name, "wb") as output:
            output.write(link.read())
    except IOError:
        print ("Unable to create %s" % name)
```
